```python
from nltk.corpus import stopwords
stop_words=set(stopwords.words("english"))
print(stop_words)

text="How are you riya ? hope you are fine"
text=re.sub('[^a-zA-Z]',' ',text)

from nltk.tokenize import word_tokenize
tokens=word_tokenize(text.lower())

filter_text=[]
for w in tokens:
  if w not in stop_words:
    filter_text.append(w)

print("Filtered Text :",filter_text)
```

```python
from nltk.stem import PorterStemmer

e_words=['wait','waits','waiting','waited']
ps=PorterStemmer()

for w in e_words:
  rootWords=ps.stem(w)
  print(rootWords)
```

```python
from nltk.stem import WordNetLemmatizer
wnl=WordNetLemmatizer()

text="wrote written fries fry"
l_tokens=nltk.tokenize.word_tokenize(text)
print(l_tokens)

for w in l_tokens:
    print("Lemma for {} is {}".format(w, wnl.lemmatize(w)))
```

```python
import pandas as pd
import math
from sklearn.feature_extraction.text  import TfidfVectorizer

doc='There is a village full of farmer who cultivate rice of basmati'

bagOfWords=doc.split(' ')
numOfWords = dict.fromkeys(bagOfWords,0)
for word in bagOfWords:
  numOfWords[word] += 1

def computeTf(wordDict,bagOfWords):
  tfDict={}
  len_count=len(bagOfWords)
  for word , count in wordDict.items():
    tfDict[word]=count/float(len_count)
  return tfDict

tf=computeTf(numOfWords,bagOfWords)
print(tf)

```

```python

docA='There is a village full of farmer who cultivate rice of basmati'

docB='There is a city with youth invest in ITs'

bagOfWordsA=docA.split(' ')
bagOfWordsB=docB.split(' ')

uniqueWords=set(bagOfWordsA).union(set(bagOfWordsB))

numOfWordsA=dict.fromkeys(uniqueWords,0)
for word in bagOfWordsA:
  numOfWordsA[word]+=1

numOfWordsB=dict.fromkeys(uniqueWords,0)
for word in bagOfWordsB:
  numOfWordsB[word]+=1

def computeIDf(documents):
  import math
  n = len(documents)
  idfDict=dict.fromkeys(documents[0].keys(),0)
  print(idfDict)

  for document in documents:
    for word ,val in document.items():
      if val>0:
        idfDict[word]+=1
  for word, val in idfDict.items():
    idfDict[word]=math.log(n/float(val))
  return idfDict

idfs=computeIDf([numOfWordsA,numOfWordsB])
print(idfs)

```
