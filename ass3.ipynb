```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data={
    'Name':['Raj','Pranjal','rajesh','Sonal'],
    'Age_group':['Adult','young','Senior','Adult'],
    'Income':[24000,50000,40000,60000]
}
df =pd.DataFrame(data)
print(df)


# Group by Age_Group and calculate summary statistics
Summary=df.groupby('Age_group')['Income'].agg(['mean','median','min','max','std'])
print(Summary)

# Create a list of Income grouped by Age Group
Income_group=df.groupby('Age_group')['Income'].apply(list)
print(Income_group)
```
