```python
# Step 1: Import required libraries
import seaborn as sns
import matplotlib.pyplot as plt

# Step 2: Load the Titanic dataset (Seaborn has it built-in)
df = sns.load_dataset('titanic')

# Step 3: Check first few rows (optional, just to understand the data)
print(df.head())

# Step 4: Plot a histogram of the 'fare' column
sns.histplot(data=df, x='fare', bins=30, kde=False)

# Step 5: Add title and labels
plt.title('Distribution of Ticket Fare')
plt.xlabel('Fare')
plt.ylabel('Number of Passengers')

# Step 6: Show the plot
plt.show()
```

### **Type of plot**:















 **Distrubition plot**

```python
#distplot
import seaborn as sns
sns.distplot(x=df['age'],bins=10,kde=True)
```

```python
#jointplot
import seaborn as sns
sns.jointplot(x='age',y='fare',data=df,kind='scatter')
```

```python
#runplot
import seaborn as sns
sns.rugplot(x='age',y='fare',data=df)
```

**catogorical plot**

```python
#barplot

import seaborn as sns
ax=sns.barplot(x='sex',y='age',data=df)
for i in ax.containers:
  ax.bar_label(i)
plt.show()
```

```python
#countplot

import seaborn as sns
ax=sns.countplot(x='sex',data=df , hue='survived', palette='rainbow')
for p in ax.containers:
  ax.bar_label(p)

plt.show()
```

```python
#boxplot
import seaborn as sns
sns.boxplot(x='sex',y='age',data=df)
```

```python

plt.figure(figsize=(6,4))
ad=sns.countplot(x=df['class'],hue=df['survived'],palette='rainbow',data=df)
for i in ad.containers:
  ad.bar_label(i)
plt.show()
```

```python
#violinplot
import seaborn as sns
sns.violinplot(x=df['class'],y='fare',data=df)
```

**Advanced plot**

```python
#stripplot
import seaborn as sns
sns.stripplot(x='sex',y='age',data=df)
```

```python
#swanplot

import seaborn as sns
sns.swarmplot(x='sex',y='age',data=df)
```

**Matrix plot**

```python
#heatmap
import seaborn as sns
import numpy as np
numerical_df=df.select_dtypes(include=np.number)
sns.heatmap(numerical_df.corr(),annot=True)
```

```python
#clustermap

import seaborn as sns
import numpy as np
numerical_df=df.select_dtypes(include=np.number)
sns.clustermap(numerical_df.corr(),annot=True)
```
