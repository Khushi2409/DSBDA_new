```python
# Step 1: Import necessary libraries
import seaborn as sns
import matplotlib.pyplot as plt

# Step 2: Load the Titanic dataset
df = sns.load_dataset('titanic')

# Step 3: Create the box plot
sns.boxplot(x='sex', y='age', hue='survived', data=df)

# Step 4: Add title and labels
plt.title('Distribution of Age by Gender and Survival Status')
plt.xlabel('Gender')
plt.ylabel('Age')

# Step 5: Show the plot
plt.show()
```

**From the box plot, it is observed that females had a significantly higher survival rate compared to males. Children (especially very young boys) also had better chances of survival. Adult males aged between 20 and 40 had the lowest survival rates. Thus, gender and age both played important roles in survival chances on the Titanic.**

**catogorical plot**

```python
#barplot

import seaborn as sns
ax=sns.barplot(x='sex',y='age',data=df)
for i in ax.containers:
  ax.bar_label(i)
plt.show()
```

```python
#countplot

import seaborn as sns
ax=sns.countplot(x='sex',data=df , hue='survived', palette='rainbow')
for p in ax.containers:
  ax.bar_label(p)

plt.show()
```

```python
first=0;
second=0;
third=0;
for i in df.pclass:
  if i==1:
    first+=1
  elif i==2:
    second+=1
  elif i==3:
    third+=1
print(first,second,third)
```

```python
first=0;
second=0;
third=0;
x = df.groupby('pclass')['survived'].sum()
print(x)
```

```python

first = 0
second = 0
third = 0

for index, row in df.iterrows():
    if row['pclass'] == 1 and row['survived'] == 1:
        first+= 1
    elif row['pclass'] == 2 and row['survived'] == 1:
        second+= 1
    elif row['pclass'] == 3 and row['survived'] == 1:
        third+= 1

print("Number of survivors in first class:", first)
print("Number of survivors in second class:", second)
print("Number of survivors in third class:", third)

# sns.barplot(x=['first class','second class','third class'],y=[first,second,third])
ax=sns.countplot(x='pclass',data=df,hue='survived')
for p in ax.containers:
  ax.bar_label(p)
plt.show()

```

```python
from matplotlib import pyplot as plt
import seaborn as sns
data=[first,second,third]
labels=['first class','second class','third class']

plt.figure(figsize=(10, 6))
plt.pie(data,labels=labels,autopct='%1.1f%%')
plt.title('Pie Chart')
plt.show()
```
